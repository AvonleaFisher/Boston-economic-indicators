## Scatterplots of Indicator Changes over Time 

The below function takes in a column from the dataframe, and the name of the column for the title and y-axis label. It returns a scatterplot which shows how values of the variable change by month and year. Month is plotted on the x-axis, and points in the plot are color-encoded by year.

```
sns.set_style("whitegrid")

def monthscatter(variable, ylab):
  plt.figure(figsize=(8,5))
  fontdict = {'fontsize':14, 'weight': 'bold', 
            'fontname': 'serif'}
  fontdictx={'fontsize': 13,
           'fontname': 'serif'}
  fontdicty={'fontsize': 13,
           'fontname': 'serif'}

  plt.title("Boston {}, Jan. 2013-Jan.2015".format(ylab), fontdict=fontdict)
  plt.xlabel("Month", fontdict=fontdicty)
  plt.ylabel(ylab, fontdict=fontdicty)
  sns.scatterplot(x=df.Month, y=variable, hue=df.Year)
```

```
monthscatter(df.unemp_rate, 'Unemployment Rate')
```

```
monthscatter(df.med_housing_price, "Median Housing Price")
```

```
monthscatter(df.logan_intl_flights, "Logan International Flights")
```

```
monthscatter(df.labor_force_part_rate, "Labor Force Participation Rate")
```

```
monthscatter(df.hotel_occup_rate, "Hotel Occup. Rate")
```
